```python
import pandas as pd
from matplotlib import pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
df = pd.read_csv('Lemonade.csv')
```

```python
df
```

```python
df.mean()
```

```python
print("Minimum Sales =",df[df['Sales'] == df['Sales'].min()]['Sales'].values[0])
print("Maximum Sales =",df[df['Sales'] == df['Sales'].max()]['Sales'].values[0])
```

```python
plt.bar(df['Sales'], df['Temperature'])
plt.show()
```

```python
plt.figure(figsize=(16,8))
plt.title("Sales and Temperature Relation")
plt.xlabel("Sales")
plt.ylabel("Temperature")
plt.plot(df['Sales'],df['Temperature'], 'ro')
```

```python
avgSales = df.groupby(['Price']).mean()['Sales']
avgSales
```

```python
plt.figure(figsize=(16,8))
plt.title("Sales and Price Relation")
plt.xlabel("Price")
plt.ylabel("Sales")
plt.bar(['0.3',' 0.5'], avgSales)
plt.show()
```

```python
days = list(df['Day'].unique())
```

```python
maxSales = df.groupby(['Day']).max()['Sales']
```

```python
plt.figure(figsize=(16,8))
plt.title("Sales and Price Relation")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.bar(days, maxSales)
plt.show()
```
